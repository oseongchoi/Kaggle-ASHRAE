```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc
import lightgbm as lgb
from sklearn.preprocessing import LabelEncoder

def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

```python
file_meter_train = 'train.csv'
file_meter_test = 'test.csv'
file_weather_train = 'weather_train.csv'
file_weather_test = 'weather_test.csv'
file_building = 'building_metadata.csv'
file_submission = 'sample_submission.csv'
```

```python
df_meter = reduce_mem_usage(pd.read_csv(file_meter_train, index_col='timestamp', parse_dates=True, infer_datetime_format=True))
df_meter.sample(5)
```

```python
df_weather = reduce_mem_usage(pd.read_csv(file_weather_train, index_col='timestamp', parse_dates=True, infer_datetime_format=True))
df_weather.sample(5)
```

```python
df_building = reduce_mem_usage(pd.read_csv(file_building))
df_building.sample(5)
```

```python
encoder = LabelEncoder()
df_building['primary_use'] = encoder.fit_transform(df_building.primary_use)
df_building = reduce_mem_usage(df_building)
df_building.sample(5)
```

```python
df = pd.merge(df_meter, df_building, how='left', on='building_id', right_index=True)
df = reduce_mem_usage(df)
df.sample(5)
```

```python
df = pd.merge(df, df_weather, how='left', on=['timestamp', 'site_id'])
df = reduce_mem_usage(df)
df.sample(5)
```

```python
df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek
df = reduce_mem_usage(df)
df.sample(5)
```

```python
# df_first = df[df.index < '2016-07-01']
# df_second = df[df.index >= '2016-07-01']
```

```python
df_first = df[df.index >= '2016-07-01']
df_second = df[df.index < '2016-07-01']
```

```python
for idx in range(4):
    print(np.log1p(df[df.meter==idx].meter_reading).quantile(0.999))
```

```python
where = (df_first.meter == 0) & (df_first.meter_reading == 0)
where |= (df_first.meter == 0) & (np.log1p(df_first.meter_reading) > 8.515)
where |= (df_first.meter == 1) & (np.log1p(df_first.meter_reading) > 9.639)
where |= (df_first.meter == 2) & (np.log1p(df_first.meter_reading) > 14.998)
where |= (df_first.meter == 3) & (np.log1p(df_first.meter_reading) > 10.786)

train_data = lgb.Dataset(
    data = df_first[~where].drop(columns='meter_reading'),
    label = np.log1p(df_first[~where].meter_reading),
    categorical_feature = ['building_id', 'meter', 'site_id', 'primary_use', 'hour', 'dayofweek']
)

booster = lgb.train({'metric': {'rmse'}}, train_data, valid_sets=[train_data], num_boost_round=200, verbose_eval=10)
```

```python
y = np.log1p(df_second.meter_reading)
X = df_second.drop(columns='meter_reading')
y_pred = booster.predict(X)
```

```python
np.sqrt(np.mean(np.power(y - y_pred, 2)))
```

```python
feature_importance = pd.DataFrame(index=df.drop(columns='meter_reading').columns, data=booster.feature_importance(), columns=['value'])
feature_importance.sort_values('value', ascending=True, inplace=True)
feature_importance.plot(kind='barh', figsize=[15, 10])
plt.show()
```

```python
df_plot = df_second[['building_id', 'meter']].copy()
df_plot['meter_true'] = y
df_plot['meter_pred'] = y_pred
```

```python
for idx, group in df_plot.groupby(['building_id', 'meter']):
    if np.random.rand() < 0.99:
        continue

    plt.figure(figsize=[25, 5])
    plt.plot(group.meter_true)
    plt.plot(group.meter_pred)
    plt.title(idx)

plt.show()
```
